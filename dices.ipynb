```python
import random
```

```python
def dice1():
    dicee1 = random.randint(1,6)
    return dicee1
```

```python
def dice2():
    dicee2 = random.randint(1,6)
    return dicee2
```

```python
def display():
    x = dice1()
    y = dice2()
    
    #d1
    
    if x == 1:
        print('The first dice shows 1')
        
    elif x == 2:
        print('The first dice shows 2')
        
    elif x == 3:
        print('The first dice shows 3')
        
    elif x == 4:
        print('The first dice shows 4')
        
    elif x == 5:
        print('The first dice shows 5')
        
    elif x == 6:
        print('The first dice shows 6')
        
    #d2
    
    if y == 1:
        print('The second dice shows 1')
        
    elif y == 2:
        print('The second dice shows 2')
        
    elif y == 3:
        print('The second dice shows 3')
        
    elif y == 4:
        print('The second dice shows 4')
        
    elif y == 5:
        print('The second dice shows 5')
        
    elif y == 6:
        print('The second dice shows 6')
        
    print('The value of this trow is: ', x+y)
    
    
         
```

```python
class player():
    
    def __init__(self):
        self.value = 0
        
    def addtrow(self):
        self.value += dice1() + dice2()
        
    def end(self):
        print('The total value of all the trows in this game is: ', self.value)
        
    def val(self):
        return self.value
```

```python
playing = True

round = 1

player1 = player()
player2 = player()

while playing:
    
    print('Welcome to dice trowing game (player1 vs player2)')
    print('Player1 goes first')
    
    ask = input('Do both players want to start a round? (y or n): ')
    
    if ask == 'y':
        print('Round: ', round)
        
        print('Player1 trows')
        dice1()
        dice2()
        display()
        player1.addtrow()
        
        print('Player2 trows')
              
        dice1()
        dice2()
        display()
        player2.addtrow()
        
        round += 1
        
    else:
        
        p1val = player1.val()
        p2val = player2.val()
        print('player one: ') 
        player1.end()
        print('player two: ') 
        player2.end()
        if p1val > p2val:
            print('Player one wins')
        else:
            print('Player two wins')
        playing == False
        break
        
    ask2 = input('Another round? (y or n): ')
    
    if ask2 == 'y':
        continue
        
    else:
        p1val = player1.val()
        p2val = player2.val()
        print('player one: ') 
        player1.end()
        print('player two: ') 
        player2.end()
        if p1val > p2val:
            print('Player one wins')
        else:
            print('Player two wins')
        playing == False
        break
    
    
```
